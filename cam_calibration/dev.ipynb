```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_object_points(object_points, plane='xy'):
    # Create a 2D plot
    fig, ax = plt.subplots(figsize=(15, 12))

    # Plot each point
    for i, point in enumerate(object_points):
        point = point * 100

        if plane == 'xy':
            ax.scatter(point[0], point[1], c='blue', s=50)
            ax.annotate(f'{i}\n({point[0]:.1f}, {point[1]:.1f}, {point[2]:.1f})', 
                        (point[0], point[1]),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=8)
        elif plane == 'xz':
            ax.scatter(point[0], point[2], c='blue', s=50)
            ax.annotate(f'{i}\n({point[0]:.1f}, {point[1]:.1f}, {point[2]:.1f})', 
                        (point[0], point[2]),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=8)

    # Set labels and title
    if plane == 'xy':
        ax.set_xlabel('X (meters)')
        ax.set_ylabel('Y (meters)') 
    elif plane == 'xz':
        ax.set_xlabel('X (meters)')
        ax.set_ylabel('Z (meters)')
    ax.set_title('Calibration Pattern Object Points')

    # Set equal aspect ratio so circles look circular
    ax.set_aspect('equal')

    # Add grid
    ax.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# lay chessboard
pattern_size = (6, 9)
square_size = 25/1000
x_offset = - (14 - 1.5 - 2.5) / 100
y_offset = - (18.5 - 2.5 - 2.5) / 100
z_offset = 1.0 / 100
object_points = np.zeros((pattern_size[0] * pattern_size[1], 3), dtype=np.float32) + z_offset
object_points[:, :2] = np.mgrid[0:pattern_size[0], 0:pattern_size[1]].T.reshape(-1, 2)[:, ::-1] * square_size + np.array([x_offset, y_offset])
plot_object_points(object_points, plane='xy')
```

```python
# stand chessboard
pattern_size = (8, 11)
square_size = 20/1000
x_offset = - (14 - 3.0 - 2.0) / 100
y_offset = + (21 - 18.5) / 100
z_offset = + (1.5 + 2.0) / 100
object_points = np.zeros((pattern_size[0] * pattern_size[1], 3), dtype=np.float32) + y_offset
object_points[:, 0::2] = np.mgrid[0:pattern_size[0], 0:pattern_size[1]].T.reshape(-1, 2)[:, ::-1] * square_size + np.array([x_offset, z_offset])
plot_object_points(object_points, plane='xz')
```
